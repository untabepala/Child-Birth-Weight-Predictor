```python
data = pd.get_dummies(data3)
```

```python
X = data.drop('BWt(kg)', axis=1)
y = data['BWt(kg)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_

best_model.score(X_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Pipeline for scaling and model
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestRegressor(random_state=42))
])

# Expanded parameter grid for Random Forest
parameters = {
    'rf__n_estimators': [50, 100, 200],
    'rf__criterion': ['squared_error', 'absolute_error'],
    'rf__max_depth': [None, 10, 20, 30],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4],
    'rf__max_features': ['sqrt', 'log2', None],
    'rf__bootstrap': [True, False]
}

# RandomizedSearchCV with cross-validation
random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=parameters, 
                                   n_iter=100, cv=5, n_jobs=-1, scoring='r2', random_state=42)
random_search.fit(X_train, y_train)

# Best model performance
best_model = random_search.best_estimator_
best_model_score = best_model.score(X_test, y_test)

print(f"Best Model Score: {best_model_score}")
print(f"Best Parameters: {random_search.best_params_}")
```

```python
def predict_weight(Bgroup, Age, Height, Parity):    
    # Check if Bgroup is in the columns
    if Bgroup not in X.columns:
        raise ValueError(f"Bgroup {Bgroup} is not a valid column name in X")

    # Get the index of Bgroup column
    loc_index = np.where(X.columns == Bgroup)[0][0]

    # Initialize the input array with zeros
    x = np.zeros(len(X.columns))
    
    # Assign the values to the input array
    x[0] = Age
    x[1] = Height
    x[2] = Parity

    # Set the value for the Bgroup column
    if loc_index >= 0:
        x[loc_index] = 1

    # Predict the weight
    return lr_clf.predict([x])[0]

# Example call
predict_weight('Bgroup_B(+VE)', 30, 153, 1)
```
